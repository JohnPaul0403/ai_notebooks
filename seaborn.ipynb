Seaborn is a library that helps to engage and create more atractive charts for data visualization. It is a complement for Matplotlib and help to create the most attractive charts.

- A powerful Python data visualization library that can help you create informative and attractive visualizations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Example

iris = sns.load_dataset("iris")
sns.histplot(data=iris)
```

```python
iris.head()
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
# Customize the scatter plot
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", size="petal_length", sizes=(20, 200), data=iris)

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot of Sepal Length vs Sepal Width ")
plt.legend(title="Species")

# Display the plot
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.lineplot(x="species", y="sepal_length", hue="species", style="species", markers=True, dashes=False, data=iris)
```

```python
sns.barplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.histplot(data=iris, x="sepal_length", bins=20, kde=True, color="green")
```

```python
sns.kdeplot(x="sepal_length", data=iris)
```

```python
sns.kdeplot(data=iris, x="sepal_length", hue="species", fill=True, alpha=0.6, linewidth=1.5)
```

```python
sns.boxplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.violinplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
# Heatmaps use color to represent the magnitude of a relationship between two categorical variables

corr = iris.corr()

# Create a heatmap of the correlation between variables
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Heatmap")
plt.show()
```

```python
sns.pairplot(data=iris)
```

```python
# Joint plot is a powerful visualization technique in seaborn that combines two different plots in one visualization
sns.jointplot(data=iris)
```

```python
# facet grids FacetGrid is a powerful seaborn tool that allows you to visualize the distribution of one variable as well as the relationship between two variables
g = sns.FacetGrid(iris, col="species")

# Plot histogram for sepal_length in each species
g.map(sns.histplot, "sepal_length")
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a bar plot to show the average total bill for each day
sns.barplot(x="day", y="total_bill", data=tips, palette="Set2")

# Add error bars to indicate confidence intervals (ci parameter)
sns.barplot(x="day", y="total_bill", data=tips, palette="Set2")

# Add a title and label the axes
plt.title("Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill")

# Show the plot
plt.show()
```

```python
data = sns.load_dataset("iris")
corr = data.corr()
sns.clustermap(corr, cmap="coolwarm")
```

```python
data = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=data)
```

```python
data = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="day", data=data, kind="scatter")
```

```python
data = sns.load_dataset("iris")
g = sns.PairGrid(data, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.show()
```
